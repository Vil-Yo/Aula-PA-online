#Atividade 1
* Somar dois números

```
#entrada
n1 = int(input("Digite o primeiro número: "))
n2 = int(input("Digite o segundo número: "))
#processamento
n3 = n1 + n2
#saida
print(f"A soma dos números é: {n3}")
```

#Atividade 2
* Subtrair dois números

```
#entrada
n1 = int(input("Digite o primeiro número: "))
n2 = int(input("Digite o segundo número: "))
#processamento
n3 = n1 - n2
#saida
print(f"A subtração dos números é: {n3}")
```

#Atividade 3
* Multiplicar dois números

```
#entrada
n1 = int(input("Digite o primeiro número: "))
n2 = int(input("Digite o segundo número: "))
#processamento
n3 = n1 * n2
#saida
print(f"A multiplicação dos números é: {n3}")
```

#Atividade 4
* Dividir dois números

```
#entrada
n1 = int(input("Digite o primeiro número: "))
n2 = int(input("Digite o segundo número: "))
#processamento
n3 = n1 / n2
#saida
print (f"A divisão dos números é: {n3}")
```

#Atividade 5
* Programar um código que realize a média aritimética de três números

```
#entrada
n1 = int(input("Digite o primeiro número: "))
n2 = int(input("Digite o segundo número: "))
n3 = int(input("Digite o terceiro número: "))
#processamento
n4 = (n1 + n2 + n3) / 3
#saida
print(f"A média dos números é: {n4:.2f}")
```

#Atividade 6
* Criar um código que receba um número inteiro e calcule:
  * o seu dobro
  * A sua metade
  * E o resto da divisão por três

```
#entrada
n1 = int(input("Digite um número inteiro: "))
#processamento
n2 = n1 * 2
n3 = n1 / 2
n4 = n1 % 3
#saida
print (f"O dobro do número é: {n2}")
print (f"A metade do número é: {n3}")
print (f"O resto da divisão por 3 é: {n4}")
```

#Atividade 7
* Criar um conversor de Graus Celsius em Fahrenheit

```
#entrada
n1 = int(input("Digite a temperatura em Celsius: "))
#processamento
n2 = (n1 * 9/5) + 32
#saida
print (f"A temperatura em Fahrenheit é: {n2:.2f}")
```

#atividade 8
* Calcular o aumento no salário de um funcionário em 10%

```
#entrada
n1 = float(input("Digite o salário do funcionário: "))
#processamento
n2 = n1 * 1.1
#saida
print (f"O novo salário do funcionário é: {n2:.2f}")
```

#atividade 9
* Calcular a área de um retângulo

```

#entrada
n1 = float(input("Digite a base do retângulo: "))
n2 = float(input("Digite a altura do retângulo: "))
#processamento
n3 = n1 * n2
#saida
print (f"A área do retângulo é: {n3:.2f}")
```

#atividade 10
* Consumo médio de combustível

```

#entrada
n1 = float(input("Digite a distância percorrida em km: "))
n2 = float(input("Digite o combustível gasto em litros: "))
#processamento
n3 = n1 / n2
#saida
print (f"o consumo médio de combustível por quilometro é: {n3:.2f}")
```
